```python
# NAME OF THE VIDEO FILES 
video_files = ['Cosmos_War_of_the_Planets.mp4', 
              'Last_man_on_earth_1964.mov', 
              'The_Gun_and_the_Pulpit.avi', 
              'The_Hill_Gang_Rides_Again.mp4', 
              'Voyage_to_the_Planet_of_Prehistoric_Women.mp4']
```

```python
# IMPORTING LIBRARIES 

from pathlib import Path
import os

text  = open("report.txt", "w") 

for video_file in video_files:
    
    requirement_satisfied = True

    text.write(f"{video_file} Analysis as follows: \n")

    ############ VIDEO FORMAT
    video_format = os.path.splitext(video_file)[1]
#     print("File extension:", video_format)
    
    if video_format != '.mp4':
        text.write("Video Format doesn't match the requirment and we needed to convert it. \n")
        requirement_satisfied = False
        
    else:
        text.write("Video Format matches the requirment! \n")
        
    ############ VIDEO CODEC
    video_codec_command = f'ffprobe -v error -select_streams v:0 -show_entries stream=codec_name -of default=nokey=1:noprint_wrappers=1 ./Videos/{video_file}'
    video_codec = !{video_codec_command}
    
    if video_codec != ['h264']:
        text.write("Video Codec doesn't match the requirment and we needed to convert it. \n")
        requirement_satisfied = False
        
    else:
        text.write("Video Codec matches the requirment! \n")
        
    ############ AUDIO CODEC 
    audio_codec_command = f'ffprobe -v error -select_streams a:0 -show_entries stream=codec_name -of default=nokey=1:noprint_wrappers=1 ./Videos/{video_file}'
    audio_codec = !{audio_codec_command}
    
    if audio_codec != ['aac']:
        text.write("Audio Codec doesn't match the requirment and we needed to convert it. \n")
        requirement_satisfied = False
        
    else:
        text.write("Audio Codec matches the requirment! \n")
    
    ############# VIDEO FPS 
    
    video_fps_command = f'ffprobe -v error -select_streams v:0 -show_entries stream=r_frame_rate -of default=nokey=1:noprint_wrappers=1 ./Videos/{video_file}'
    video_fps = !{video_fps_command}
    
    if video_fps != ['25/1']:
        text.write("Video Frame Rate doesn't match the requirment and we needed to convert it. \n")
        requirement_satisfied = False
        
    else:
        text.write("Video Frame Rate matches the requirment! \n")
    
    ############ VIDEO ASPECT RATIO
    
    video_aspect_command = f'ffprobe -v error -select_streams v:0 -show_entries stream=display_aspect_ratio -of default=nokey=1:noprint_wrappers=1 ./Videos/{video_file}'
    video_aspect = !{video_aspect_command}
    
    if video_aspect != ['16:9']:
        text.write("Video Aspect Ratio doesn't match the requirment and we needed to convert it. \n")
        requirement_satisfied = False
        
    else:
        text.write("Video Aspect Ratio matches the requirment! \n")
    
    ########### VIDEO RESOLUTION
    
    video_res_command = f'ffprobe -v error -select_streams v:0 -show_entries stream=width,height -of default=nokey=1:noprint_wrappers=1 ./Videos/{video_file}'
    video_res = !{video_res_command}
    
    if video_res != ['640', '360']:
        text.write("Video Resolution doesn't match the requirment and we needed to convert it. \n")
        requirement_satisfied = False
        
    else:
        text.write("Video Resolution matches the requirment! \n")
    
    ########## VIDEO BIT RATE
    
    video_bitrate_command = f'ffprobe -v error -select_streams v:0 -show_entries stream=bit_rate -of default=nokey=1:noprint_wrappers=1 ./Videos/{video_file}'
    video_bitrate = !{video_bitrate_command}
    
    if int(video_bitrate[0]) < 2000000 or int(video_bitrate[0]) > 5000000:
        text.write("Video Bit Rate doesn't match the requirment and we needed to convert it. \n")
        requirement_satisfied = False
        
    else:
        text.write("Video Bit Rate matches the requirment! \n")
    
    ######## AUDIO BIT RATE 
    
    audio_bitrate_command = f'ffprobe -v error -select_streams a:0 -show_entries stream=bit_rate -of default=nokey=1:noprint_wrappers=1 ./Videos/{video_file}'
    audio_bitrate = !{audio_bitrate_command}
    
    if int(audio_bitrate[0]) > 256000 :
        text.write("Audio Bit Rate doesn't match the requirment and we needed to convert it. \n")
        requirement_satisfied = False
        
    else:
        text.write("Audio Bit Rate matches the requirment! \n")
        
    ######## AUDIO CHANNELS 
    
    audio_channels_command = f'ffprobe -v error -select_streams a:0 -show_entries stream=channels -of default=nokey=1:noprint_wrappers=1 ./Videos/{video_file}'
    audio_channels = !{audio_channels_command}
    
    if audio_channels != ['2'] :
        text.write("Audio Channels doesn't match the requirment and we needed to convert it. \n")
        requirement_satisfied = False
        
    else:
        text.write("Audio Channels matches the requirment! \n")
        
    
    # IF THE REQUIREMENTS DONT MATCH , THEN CREATE THE REQUIRED VERSION WITH OKFORMAT OF THE VIDEO 
    if requirement_satisfied == False:
        conversion_command = f'ffmpeg -i ./Videos/{video_file} -c:v libx264 -vf "fps=25, scale=640:360, setsar=16:9" -b:v 2M -c:a aac -b:a 256k -ac 2 -strict experimental ./Videos/{video_file}_formatOK.mp4'
        formatOK = !{conversion_command}
    
    text.write("\n \n")

text.close() 
```
